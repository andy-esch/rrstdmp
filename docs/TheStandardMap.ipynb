# The Standard Map

The standard map is a two-dimensional representation of the position and momentum of a kicked rotor. The Hamiltonian is given by:
<div style="margin: 20px;">
$$
\large
\begin{equation}
H\left[ x(t),p(t)\right] = \frac{p^2}{2m} + k\cos\left(x\right)\delta\left(t\right)
\end{equation}
$$
</div>
As you can see, the energy is gained by periodic kicking of strength $k$ at different angles $x$. The rotor can be a bead on a circular ring, a ball constrained to roller in a circle on a flat table, or bar that rotates around a central point. In all cases, the only force acting on the particle is the periodic kick with direction given by 
<div style="margin: 20px;">
$$
\large
    -\frac{\partial\hphantom{x}}{\partial x} \cos\left(x\right) = \sin\left(x\right)
$$
</div>
Applying Hamilton's Equations to this functional gives the following relationship:
<div style="margin: 20px;">
$$
\large
\begin{align}
p^{\prime} &= p - \frac{k}{2\pi}\sin\left(2\pi x\right)\newline
x^{\prime} &= x + p^{\prime}
\end{align}
$$
</div>

In Python, these equations look as follows.

Note: Since each value depends on the ones that precede it, I have not found a very pythonic way to populate the map without doing for loops.

```python
def stdmp(x0,y0,n=100,k=0.97163540631):
    """
        Implementation of the Chirikov standard map

        Parameters:
        -----------
        x0 : float
            initial x position
        y0 : float 
            initial y position
        n : int, optional
            number of (x,y) pairs in the output map, defaults at 100 iterations
        k : float, optional
            kicking strength parameter, defaults at Greene's constant
    
    """
    twopi = 2. * np.pi
    x = np.empty(n,dtype='double')
    y = np.empty(n,dtype='double')
    x[0] = x0
    y[0] = y0

    for i in range(1,n-1):
        y[i] = np.fmod(y[i-1] - k / (twopi) * np.sin(twopi * x[i-1]) + 1.0,1.0)
        x[i] = np.fmod(x[i-1] + y[i] + 1.0,1.0)
    
    return x, y
```

```python
n_points = 800
n_entries = 30
k = 0.97163540631
a = np.empty(n_points)
b = np.empty(n_points)

# Plot trajectories of standard map
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.1, 0.1, 1.0, 1.0])
for j in range(n_entries):
    a, b = stdmp(np.random.rand(1),np.random.rand(1),n=n_points,k=k)
    ax.plot(a,b,'.',markersize=1)
ax.axis((0.,1.,0.,1.))
ax.set_xlabel('$x$')
ax.set_ylabel('$p$')
ax.set_title('Phase Space of the Standard Map\n for $k= %f$' % k)
matplotlib.rcParams.update({'font.size': 18, 'text.usetex': True, 'font.family': 'serif'});

del a, b, ax, fig, k, j
```

```python
k_sample = np.array([0.0, 0.5, 0.97163540631, 5.0])
n_points = 400
n_entries = 25
k = 0.97163540631
a = np.empty(n_points)
b = np.empty(n_points)

fig, axs = plt.subplots(nrows=2, ncols=2, sharex='col', sharey='row')
fig.set_size_inches(10,10)
# Text box properties
props = dict(boxstyle='round', facecolor='white', edgecolor='black')
fig.suptitle('Standard Map at various values of $k$',fontsize=24)
i = 0
for x in [item for sublist in axs for item in sublist]:
    for j in range(n_entries):
        a, b = stdmp(np.random.rand(1),np.random.rand(1),n=n_points,k=k_sample[i])
        x.plot(a,b,'.',markersize=1)
    if i == 1:
        "Plot golden torus"
        a, b = stdmp(0.0,0.664758,k=k_sample[i],n=n_points)
        x.plot(a,b,'k.',markersize=2)
        x.plot(a[0],b[0],'o',ms=15,mfc='blue',mec='red',mew=3,alpha=0.5)
        
        a, b = stdmp(0.0,0.335242,k=k_sample[i],n=n_points)
        x.plot(a,b,'k.',markersize=2)
        x.plot(a[0],b[0],'o',ms=15,mfc='blue',mec='red',mew=3,alpha=0.5)
    x.axis((0.,1.,0.,1.))
    if (i > 1):
        x.set_xlabel('$x$')
    if (i % 2 == 0):
        x.set_ylabel('$p$')
    if (i == 1):
        strtext = "$k \\approx %3.3f$" % k_sample[i]
    else:
        strtext = "$k = %3.3f$" % k_sample[i]
    x.text(0.04,0.05,strtext, bbox=props)
#    x.set_title('Phase Space of the Standard Map\n for $k= %f$' % k_sample[i])
    i += 1
    x.plot()

# Get rid of variables no longer of use
del i, a, b, x, fig, axs, j
```
